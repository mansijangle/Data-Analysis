```python
ax = sns.countplot(x='Gender',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# plot the group which is highest gender to spend a lot of money in shopping 
```

```python
df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
sale_gem = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Gender',y='Amount',data = sale_gem)
```

```python
# above graph female is spending a amonunt of money for shopping in given data set
```

```python
ax = sns.countplot(data = df,x='Age Group',hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# from above graph most of the buyer is in between the 25-36 years of female
```

```python
sale_gem = df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Age Group',y='Amount',data = sale_gem)
```

```python
sale_state = df.groupby(['State'],as_index=False)['Orders'].sum().sort_values(by='Orders',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='State',y='Orders',data = sale_state)
```

```python

# from above graph we can see that most of the order from the maharashtra and up
```

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(x='Occupation',data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sale_gem = df.groupby(['Occupation'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)

sns.barplot(x='Occupation',y='Amount',data = sale_gem)
```

```python
# from above graph we can see that IT sector purshaed more than other sector
```

```python
sale_gem = df.groupby(['Product_Category'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Product_Category',y='Amount',data = sale_gem)
```
